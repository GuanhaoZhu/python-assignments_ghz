```python
#importing the library as shortname 
import matplotlib.pyplot as plt 
```

```python
nailpolish_inventory = {
    'red': 5.555, 
    'pink': 1, 
    'black': 2, 
    'glitter': 4, 
    'green': 3, 
    'neutral': 7,
    'blue': 5, 
    }

len(nailpolish_inventory)
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.bar(range(len(nailpolish_inventory)), list(nailpolish_inventory.values()), align='center', color=['#CE3D1E', '#EA42A2', '#080507', '#FFFFE7', '#BEF2C3', '#CFCF97', '#97B5CF'])
plt.xticks(range(len(nailpolish_inventory)), list(nailpolish_inventory.keys()), rotation = 45)
plt.show()
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter
plt.scatter(range(len(nailpolish_inventory)), list(nailpolish_inventory.values()), color=['#CE3D1E', '#EA42A2', '#080507', '#FFFFE7', '#BEF2C3', '#CFCF97', '#97B5CF'])
plt.xticks(range(len(nailpolish_inventory)), list(nailpolish_inventory.keys()), rotation = 45)
plt.show()
```
