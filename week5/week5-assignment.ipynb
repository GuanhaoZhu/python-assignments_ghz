```python
a = []

for i in range(len(records)):
    lst = records[i]['comName'],records[i]['howMany']
    for keys in lst:
        a.append(keys)
print(a)
```

```python
def Convert(lst):
    res_dct = {lst[i]: lst[i + 1] for i in range(0, len(lst), 2)}
    return res_dct
         
# Driver code
lst = a
print(Convert(lst))

import matplotlib.pyplot as plt 
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
bird_list = Convert(lst)

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

plt.figure(figsize=(15,6))
plt.bar(range(len(bird_list)), list(bird_list.values()), align='center')
plt.xticks(range(len(bird_list)), list(bird_list.keys()), rotation = 90)
plt.show()
```
